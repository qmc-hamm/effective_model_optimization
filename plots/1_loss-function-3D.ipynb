```python
fix, ax = plt.subplots(nrows=1, ncols=2, sharey=True, gridspec_kw={'width_ratios':[3, 1]})

N = 10 # state cutoff

ax[0].scatter(np.ones(N), np.arange(N)) # Model 1
ax[0].scatter(2*np.ones(N), np.arange(N)) # Model 2
ax[0].scatter(3*np.ones(N), np.arange(N)) # Model 3
ax[0].set_xticks([1,2,3], labels=["model1","model2", "model3"])

ax[1].scatter(np.ones(N), np.arange(N)) # Ab Initio data
ax[1].set_xticks([1], labels=["ab initio"])

ax[0].set_ylabel("Energy (eV)")
ax[0].set_xlabel("Model")
ax[1].set_xlabel("ab initio data")
```

```python
fix, ax = plt.subplots(nrows=1, ncols=2, sharey=True, gridspec_kw={'width_ratios':[3, 1]})

N = 10 # state cutoff
cmap = 'Blues_r'
vmin = 0
vmax = 1.2

# c is some relavant descriptor
ax[0].scatter(np.ones(N), np.arange(N), c=0.1*np.arange(N), cmap=cmap, vmin=vmin, vmax=vmax) # Model 1
ax[0].scatter(2*np.ones(N), np.arange(N), c=0.8*0.1*np.arange(N), cmap=cmap, vmin=vmin, vmax=vmax) # Model 2
ax[0].scatter(3*np.ones(N), np.arange(N), c=0.6*0.1*np.arange(N), cmap=cmap, vmin=vmin, vmax=vmax) # Model 3
ax[0].set_xticks([1,2,3], labels=["model1","model2", "model3"])

ax[1].scatter(np.ones(N), np.arange(N), c=0.1*np.arange(N), cmap=cmap, vmin=vmin, vmax=vmax) # Ab Initio data
ax[1].set_xticks([1], labels=["ab initio"])

ax[0].set_ylabel("Energy (eV)")
ax[0].set_xlabel("Model")
ax[1].set_xlabel("ab initio data")
```

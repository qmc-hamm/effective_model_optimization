```python
def entropy(boltzmann_weights):
    return -np.sum(boltzmann_weights*np.log(boltzmann_weights))

T = np.linspace(1, 100, 100)
S = np.asarray([entropy(compute_boltzmann_weights(df, t)) for t in T])
plt.plot(T, S)
plt.xlabel("Temperature (eV)")
plt.ylabel("Entropy")
plt.figure()
effective_number = np.exp(S)
plt.plot(T, effective_number)
plt.xlabel("Temperature (eV)")
plt.ylabel("Effective number of states")
```
